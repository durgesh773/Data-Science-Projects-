```python
dftr=pd.get_dummies(df_tr,prefix=None,prefix_sep='_',drop_first=True)
```

```python
dftr
```

```python
dfte=pd.get_dummies(df_te,prefix=None,prefix_sep='_',drop_first=True)
```

```python
dfte
```

```python
xtr=dftr.iloc[:,:42]
ytr=dftr.iloc[:,42]
```

```python
xte=dfte.iloc[:,:42]
yte=dfte.iloc[:,42]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(xtr)
X_train = scaler.transform(xtr)
X_test = scaler.transform(xte)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 2)
classifier.fit(X_train, ytr)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(yte, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(yte, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(yte,y_pred)
print("Accuracy:",result2)
```

```python
test = pd.read_csv('test.csv',sep=';')
train = pd.read_csv('train.csv',sep=';')

df = pd.concat([test, train])
```

```python
df.head()
```

```python
df.info()
```

```python
df1 = pd.get_dummies(df,prefix=None,prefix_sep='_',drop_first=True)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.drop('y_yes',axis=1), 
                                                    df1['y_yes'], test_size=0.10, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, predictions)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,predictions)
print("Accuracy:",result2)
```
